# **Práticas recomendadas**:

- **Encapsulamento**: Ocultar os detalhes internos de uma classe e fornecer uma interface consistente para interagir com ela. Isso é alcançado através da definição de atributos privados e do uso de getters e setters para acessar e modificar esses atributos.

- **Abstração**: Representar conceitos do mundo real como objetos com características (atributos) e comportamentos (métodos). A abstração permite que você modele e trabalhe com entidades relevantes para o domínio do problema.

- **Herança**: Permite que uma classe herde atributos e comportamentos de uma classe base (superclasse). A herança promove a reutilização de código e a criação de uma hierarquia de classes.

- **Polimorfismo**: Capacidade de uma classe ser tratada como uma instância de sua classe base, além de poder assumir diferentes formas. O polimorfismo permite escrever código genérico que pode lidar com diferentes tipos de objetos, facilitando a extensibilidade e a flexibilidade do código.


# **Princípios Recomendadas**:

- **Princípio da Responsabilidade Única** (SRP): Uma classe deve ter uma única responsabilidade e um único motivo para mudar. Cada classe deve ser focada em uma única tarefa ou responsabilidade.

- **Princípio Aberto/Fechado** (OCP): As classes devem estar abertas para extensão, mas fechadas para modificação. Isso significa que você pode adicionar novas funcionalidades ou comportamentos às classes existentes sem modificar seu código fonte.

- **Princípio da Inversão de Dependência** (DIP): Depender de abstrações, não de implementações concretas. Isso envolve programar para interfaces ou classes abstratas em vez de classes concretas, permitindo maior flexibilidade e facilitando a substituição de implementações.

- **Princípio da Substituição de Liskov** (LSP): As classes derivadas devem ser substituíveis por suas classes base. Isso significa que uma classe derivada deve ser capaz de substituir sua classe base sem alterar o comportamento correto do programa.

- **Princípio da Segregação de Interface** (ISP): Clientes não devem ser forçados a depender de interfaces que não usam. É melhor ter interfaces mais específicas do que uma única interface grande.

- **Princípio do Mínimo Conhecimento** (ou Lei de Demeter): Uma classe deve ter conhecimento limitado sobre outras classes. Cada objeto deve se comunicar apenas com seus vizinhos imediatos e não deve conhecer detalhes internos de objetos relacionados indiretamente.

# **Modificadores de Acesso:**

Na Programação Orientada a Objetos (POO), os modificadores de acesso são palavras-chave que definem o nível de acesso aos membros (atributos e métodos) de uma classe. Esses **modificadores controlam a visibilidade e a acessibilidade dos membros em relação a outras classes e objetos**.

- Em Java, existem quatro modificadores de acesso principais:

  - **public**: O membro é acessível por qualquer classe em qualquer pacote. Não há restrições de acesso.

  - **private**: O membro é acessível apenas dentro da própria classe. Ele não é visível nem acessível por outras classes, mesmo que sejam do mesmo pacote.

  - **protected**: O membro é acessível dentro da própria classe, por classes do mesmo pacote e por subclasses (classes filhas) mesmo que estejam em pacotes diferentes.

 - **default** (ausência de modificador de acesso): O membro é acessível dentro da própria classe e por classes do mesmo pacote. Se nenhum modificador de acesso for especificado, o membro terá acesso padrão.

Lembre sempre sobre o contexto dos atributos dentro da classe, dados sensíveis precisão de modificadores como private ou protected depedendo do contexto.

# **Getters / Setters:**

Getters e setters são métodos utilizados para acessar e modificar os atributos de uma classe, seguindo o princípio de encapsulamento da Programação Orientada a Objetos (POO). Eles fornecem uma interface controlada e consistente para interagir com os atributos de uma classe, garantindo a segurança e o controle sobre os dados.

Um getter é um método que permite obter o valor de um atributo, enquanto um setter é um método que permite modificar o valor de um atributo. Os getters e setters seguem uma convenção de nomenclatura, onde o nome do getter começa com "get" seguido pelo nome do atributo (em camel case), e o nome do setter começa com "set" seguido pelo nome do atributo (em camel case) e recebe um parâmetro com o novo valor a ser atribuído.

# **Sobrecargas**:

# - **Sobrecarga de métodos**:

  - A sobrecarga de métodos ocorre quando uma classe tem vários métodos com o mesmo nome, mas com diferentes parâmetros (tipo, número ou ordem).

  - Os métodos sobrecarregados podem ter tipos de retorno diferentes, mas a assinatura do método (nome e parâmetros) precisa ser diferente.

  - A sobrecarga de métodos permite que você forneça diferentes formas de um método com comportamentos específicos, mas relacionados, em uma classe.

  - É útil quando você deseja realizar operações semelhantes, mas com diferentes tipos de entrada ou comportamentos.

#- **Sobrecarga de construtores**:

  - A sobrecarga de construtores ocorre quando uma classe tem vários construtores com diferentes parâmetros.

  - Os construtores sobrecarregados têm o mesmo nome da classe, mas diferentes listas de parâmetros.

  - Os construtores sobrecarregados são usados para criar objetos da classe com diferentes conjuntos de **valores iniciais** para seus atributos.

  - É útil quando você deseja fornecer diferentes formas de inicialização para um objeto da classe.

# **Acoplamento**:

O acoplamento na Programação Orientada a Objetos (POO) refere-se ao grau de dependência entre as classes e componentes de um sistema. 

- Um alto acoplamento indica que as classes estão fortemente interligadas, o que pode tornar o código mais difícil de entender, manter e modificar.

- Um baixo acoplamento significa que as classes têm dependências limitadas e são mais independentes (uma classe deve saber o mínimo possível sobre outras classes).
